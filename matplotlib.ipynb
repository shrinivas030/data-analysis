```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,20)
y=x*x
z=x+y
```

```python
plt.plot(x,y,z,x)
```

```python
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sample')
```

```python
plt.subplot(3,2,1)
plt.plot(x,y)
plt.subplot(3,2,2)
plt.plot(x,z)
plt.subplot(3,2,3)
plt.plot(y,x)
plt.subplot(3,2,4)
plt.plot(y,z)
plt.subplot(3,2,5)
plt.plot(z,x)
plt.subplot(3,2,6)
plt.plot(z,y)
plt.tight_layout()
```
