```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.head(10)
```

```python
sns.rugplot(iris['sepal_width'])
```

```python
sns.kdeplot(iris['sepal_length'])
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.residplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='day')
```

```python
x=sns.pairplot(tips,hue='sex')
x.map_diag(sns.distplot)
x.map_upper(sns.kdeplot)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.pointplot(x='total_bill',y='day',data=tips)
```

```python
sns.factorplot(x='total_bill',y='day',data=tips)
```

```python
sns.swarmplot(x='smoker',y='tip',data=tips,hue='sex')
```

```python
sns.stripplot(x='smoker',y='tip',data=tips,hue='sex',palette='cool')
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
tip2=tips.corr()
```

```python
sns.heatmap(tip2,annot=True,linewidths=10,linecolor='black')
```
